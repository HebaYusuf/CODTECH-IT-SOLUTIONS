```python
import numpy as np

class RecommendationSystem:
    def __init__(self, num_users, num_items):
        self.num_users = num_users
        self.num_items = num_items
        self.user_item_matrix = np.zeros((num_users, num_items))
        self.item_similarity_matrix = None

    def add_rating(self, user_id, item_id, rating):
        self.user_item_matrix[user_id, item_id] = rating

    def calculate_similarity(self):
        # Calculate item-item similarity using cosine similarity
        self.item_similarity_matrix = np.dot(self.user_item_matrix.T, self.user_item_matrix)
        norms = np.linalg.norm(self.user_item_matrix, axis=0)

        # Avoid division by zero by checking for zero norms
        non_zero_norms = norms > 0
        self.item_similarity_matrix[:, non_zero_norms] /= np.outer(norms[non_zero_norms], norms[non_zero_norms])

    def recommend_items(self, user_id, num_recommendations=5):
        if self.item_similarity_matrix is None:
            self.calculate_similarity()

        user_preferences = self.user_item_matrix[user_id]
        item_scores = np.dot(self.item_similarity_matrix, user_preferences)

        # Exclude items already rated by the user
        item_scores[user_preferences != 0] = -1

        # Get indices of top recommendations
        recommended_item_indices = np.argsort(item_scores)[::-1][:num_recommendations]

        return recommended_item_indices

# Example usage:
if __name__ == "__main__":
    # Create a recommendation system with 5 users and 10 items
    recommendation_system = RecommendationSystem(5, 10)

    # Add some ratings
    recommendation_system.add_rating(0, 0, 5)
    recommendation_system.add_rating(0, 1, 4)
    recommendation_system.add_rating(1, 1, 5)
    recommendation_system.add_rating(2, 2, 3)
    recommendation_system.add_rating(3, 3, 4)
    recommendation_system.add_rating(4, 4, 5)

    # Recommend items for user 0
    recommended_items = recommendation_system.recommend_items(0)

    print("User 0's recommended items:")
    for item_index in recommended_items:
        print(f"Item {item_index}")

```
